```python
df['alkali_metals'] = df[['H', 'Li', 'Na', 'K', 'Rb', 'Cs']].sum(axis=1)
df['alkaline_earth_metals'] = df[['Be', 'Mg', 'Ca', 'Sr', 'Ba']].sum(axis=1)
df['transition_metals'] = df[['Sc', 'Ti', 'V', 'Cr', 'Mn', 'Fe', 'Co', 'Ni', 'Cu', 'Zn', 'Y', 'Zr', 'Nb', 'Mo', 'Tc', 'Ru', 'Rh', 'Pd', 'Ag', 'Cd', 'In', 'Sn', 'Hf', 'Ta', 'W', 'Re', 'Os', 'Ir', 'Pt', 'Au', 'Hg']].sum(axis=1)
df['post_transition_metals'] = df[['Al', 'Ga', 'Ge', 'In', 'Sn', 'Tl', 'Pb', 'Bi']].sum(axis=1)
df['metalloids'] = df[['B', 'Si', 'As', 'Sb', 'Te']].sum(axis=1)
df['non_metals'] = df[['C', 'N', 'O', 'F', 'P', 'S', 'Cl', 'Se', 'Br', 'I']].sum(axis=1)
df['noble_gases'] = df[['He', 'Ne', 'Ar', 'Kr', 'Xe', 'Rn']].sum(axis=1)
df['lanthanides'] = df[['La', 'Ce', 'Pr', 'Nd', 'Pm', 'Sm', 'Eu', 'Gd', 'Tb', 'Dy', 'Ho', 'Er', 'Tm', 'Yb', 'Lu']].sum(axis=1)
```
